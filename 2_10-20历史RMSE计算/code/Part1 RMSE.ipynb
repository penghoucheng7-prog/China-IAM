```python
import pandas as pd

# ä½¿ç¨åå§å­ç¬¦ä¸²é¿åè·¯å¾é®é¢
file_path = r'C:\Users\phc\Desktop\AI-IAMs-China\Adjusted_data\Historical_error\RMSE_data\error_metrics_avg_4models.csv'

# è¯»å CSV æä»¶
data = pd.read_csv(file_path)

# æ¾ç¤ºåå è¡æ°æ®ï¼äºè§£æ°æ®ç»æ
data.head()
```

```python
# Calculate sample standard deviation for each 'Variable' in the 'NORMAL_bias' column
std_per_variable = data.groupby('Variable')['NORMAL_bias'].std()

# Function to classify data based on the calculated standard deviation
def classify_data(row, std_dev):
    lower_bound = -0.5 * std_dev
    upper_bound = 0.5 * std_dev
    if row['NORMAL_bias'] < lower_bound:
        return 'Low'
    elif row['NORMAL_bias'] > upper_bound:
        return 'High'
    else:
        return 'Medium'

# Apply classification to each row based on the standard deviation of its 'Variable'
data['Classification'] = data.apply(lambda row: classify_data(row, std_per_variable[row['Variable']]), axis=1)

# Display updated dataframe to check the classification
data[['Variable', 'NORMAL_bias', 'Classification']].head()

# Save the updated dataframe to a new Excel file
updated_file_path = r'C:\Users\phc\Desktop\AI-IAMs-China\Adjusted_data\Historical_error\RMSE_data\rmse_results_models_average.csv'
data.to_csv(updated_file_path, index=False)
```
